# finding the unique value 

```python
car_data['Fuel_Type'].unique()
```

# endcoding unique value from feature 

```python
car_data['Fuel_Type']=car_data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
car_data['Seller_Type']=car_data['Seller_Type'].map({'Dealer':0,'Individual':1})
```

```python
car_data['Transmission']=car_data['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
car_data
```

```python
data=car_data.drop('Car_Name',axis=1)
```

```python
data
```

```python
# feature encoding successfully done 
```

```python
# splitting feature and target feature  
```

```python
x=data[['Year','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner']]
```

```python
x
```

```python
y=data['Selling_Price']
```

```python
y
```

```python
# import train_test
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
data.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
```

```python
model = lm.fit(X_train,Y_train)
```

```python
model
```

```python
y_pred= model.predict(X_train)
```

```python
model.score(X_test,Y_test)
```

```python
model.coef_
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,Y_train)
```
